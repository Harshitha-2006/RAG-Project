```python
from langchain_community.document_loaders import DirectoryLoader,TextLoader
folder_path="RAG_DATA"
loader=DirectoryLoader(
    folder_path,
    glob="**/*.txt",
    loader_cls=TextLoader
)
docs=loader.load()
for i,d in enumerate(docs):
    print(f"\n--- Document {i+1} ---")
    print(d.page_content[:500])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splits=text_splitter.split_documents(docs)
```

```python
for i in range(len(splits)):
  print(splits[i])
```

```python
print(len(splits))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain.vectorstores import Chroma
vectorstore=Chroma.from_documents(documents=splits, embedding=embedding)
```

```python
retriver=vectorstore.as_retriever()
```

```python
!pip install transformers accelerate sentencepiece
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline
generator=pipeline("text2text-generation", model="google/flan-t5-base")
llm=HuggingFacePipeline(pipeline=generator)
```

```python
from langchain.prompts import PromptTemplate
rag_prompt=PromptTemplate(
    input_variables=["context","question"],
    template="""
You are an assistant for answering questions about college rules, fees, hostel, facilities, and regulations.
Answer the question based ONLY on the context provided below.
If the context does not contain the answer, say "I don't know".
Rules for answering:
- Be concise and direct.
- Give at most **one or two sentences**.
- Stop at the first full stop that contains the answer.
- Do NOT include extra details from the context.
Context:
{context}
Question:
{question}
Answer:
"""
)
```

```python
from langchain.chains import RetrievalQA
rag_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriver,
    chain_type="stuff",
    chain_type_kwargs={"prompt":rag_prompt}
)
```

```python
def shorten_answer(answer:dict) -> str:
    answer_text=answer.get('result', '')
    answer_text=answer_text.strip()
    if answer_text.lower().startswith("yes"):
        return "Yes, cell phones are allowed in the library."
    elif answer_text.lower().startswith("no"):
        return "No, cell phones are not allowed in the library."
    parts=answer_text.split(".")
    if parts and parts[0].strip():
        return parts[0].strip() + "."
    return answer_text
```

```python
def ask(question: str) -> str:
    raw=rag_chain.invoke(question)
    return raw["result"].strip()
```

```python
print(ask("What is transportation fee per annum?"))
```

```python
print(ask("Is Wi-Fi available in hostel"))
```

```python
print(ask("what are facilities available in hostel?"))
```

```python
print(ask("what are library timings?"))
```

```python
print(ask("what is the salary of teachers"))
```

```python
raw=rag_chain.invoke("Are cell phones allowed in the library?")
final=shorten_answer(raw)
print(final)
```

```python
raw=rag_chain.invoke("What types of books are accepted by library as donation?")
final=shorten_answer(raw)
print(final)
```
